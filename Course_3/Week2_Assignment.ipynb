### Question 2

Write a function that fits a polynomial LinearRegression model on the training data `X_train` for degrees 0 through 9. For each model compute the $R^2$ (coefficient of determination) regression score on the training data as well as the the test data, and return both of these arrays in a tuple.

*This function should return one tuple of numpy arrays `(r2_train, r2_test)`. Both arrays should have shape `(10,)`*

```python
def answer_two():
    from sklearn.linear_model import LinearRegression
    from sklearn.preprocessing import PolynomialFeatures
    from sklearn.metrics.regression import r2_score
    #First of all, initialize two empty arrays to hold r2_train and r2_test
    r2_train = np.empty(10)
    r2_test = np.empty(10)
    for index, deg in enumerate(range(10)):
        poly = PolynomialFeatures(degree=deg)
        
        '''Scores on training dataset'''
        #We generate polynomial features for X_train
        X_train_poly = poly.fit_transform(X_train.reshape(11,1))
        #Fit the data to the LinearRegression model
        linear_regression = LinearRegression().fit(X_train_poly, y_train)
        #Store the score to the index's index of r2_train array
        r2_train[index] = linear_regression.score(X_train_poly, y_train)
        
        '''Scores on test dataset'''
        #We generate polynomial features for X_test
        X_test_poly = poly.fit_transform(X_test.reshape(4,1))
        #Store the score to the index's index of r2_test array
        r2_test[index] = linear_regression.score(X_test_poly, y_test)
        #print(len(r2_train))
        #print(len(r2_test))
    return tuple((r2_train, r2_test))
answer_two()
```

### Question 3

Based on the $R^2$ scores from question 2 (degree levels 0 through 9), what degree level corresponds to a model that is underfitting? What degree level corresponds to a model that is overfitting? What choice of degree level would provide a model with good generalization performance on this dataset? Note: there may be multiple correct solutions to this question.

(Hint: Try plotting the $R^2$ scores from question 2 to visualize the relationship between degree level and $R^2$)

*This function should return one tuple with the degree values in this order: `(Underfitting, Overfitting, Good_Generalization)`*

```python
def answer_three():
    
    # Your code here
    
    #We create two Series objects separating the test and train scores 
    r2_scores_train = pd.Series(answer_two()[0])
    r2_scores_test = pd.Series(answer_two()[1])
    difference = r2_scores_train - r2_scores_test
    
    difference.sort_values(inplace=True, ascending=True)
    good_generalization = difference.index[0]
    
    difference.sort_values(inplace=True, ascending=False)
    overfitting = difference.index[0]
    
    underfitting = r2_scores_train.sort_values(ascending=True).index[0]
    return tuple((underfitting, overfitting, good_generalization))

answer_three()
```

For this question, train two models: a non-regularized LinearRegression model (default parameters) and a regularized Lasso Regression model (with parameters `alpha=0.01`, `max_iter=10000`) on polynomial features of degree 12. ### Question 5

Using `X_train2` and `y_train2` from the preceeding cell, train a DecisionTreeClassifier with default parameters and random_state=0. What are the 5 most important features found by the decision tree?

As a reminder, the feature names are available in the `X_train2.columns` property, and the order of the features in `X_train2.columns` matches the order of the feature importance values in the classifier's `feature_importances_` property. 

*This function should return a list of length 5 containing the feature names in descending order of importance.*


```python
def answer_five():
    from sklearn.tree import DecisionTreeClassifier
    
    #First of all, we train the Decision Tree Classifier with train datas X_train2, y_train2
    decision_tree_clf = DecisionTreeClassifier(random_state=0).fit(X_train2, y_train2)
    #then, I create a dataframe with 2 columns that store Feature name and Importance correspondingly
    features = pd.DataFrame({'Feature Name': X_train2.columns.values,
                             'Importance': decision_tree_clf.feature_importances_})
    #afterwards, sort the features dataframe by the column 'Importance'
    features = features.sort_values(by='Importance', ascending=False)
    #finally I return the nlargest Importance, (n=5)
    return features.nlargest(5,'Importance')['Feature Name'].tolist()
answer_five()
```
