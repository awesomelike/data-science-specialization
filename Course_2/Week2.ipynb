# Line plots

```python
import numpy as np
```

```python
lin_data = np.array([1,2,3,4,5,6,7,8])
quadratic_data = lin_data**2

plt.figure()
plt.plot(lin_data,'-o',quadratic_data,'-o')
```

```python
plt.plot([20,30,50],'--r')
```

```python
plt.xlabel('X Data')
plt.ylabel('Y Data')
plt.title('The title')
plt.legend(['Baseline','Competition','Us'])
```

```python
plt.gca().fill_between(range(len(lin_data)),
                            lin_data,
                            quadratic_data,
                            facecolor='yellow',
                            alpha=0.2)
```

```python
plt.figure()
```

```python
obs_data = np.arange('2019-05-01','2019-05-09',dtype='datetime64[D]')
plt.plot(obs_data,lin_data,'-o',
        obs_data,quadratic_data,'-o')
```

```python
x=plt.gca().xaxis

for i in x.get_ticklabels():
    i.set_rotation(45)
plt.subplots_adjust(bottom=0.25)
```

```python
ax = plt.gca()
ax.set_xlabel('Date')
ax.set_ylabel('Units')
ax.set_title('Quadratic x Linear')
```

```python
ax.set_title('Quadratic ($x^2$) x Linear ($x$) performance.')
```
